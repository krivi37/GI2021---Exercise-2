```python
import pysam
samfile = pysam.AlignmentFile("merged-tumor.bam", "rb")
first_read = next(samfile.head(1, multiple_iterators=False))
print('First read fields:' + str(first_read))
print('First read flag: ' + str(first_read.flag))
some_read_flags = [read.flag for read in samfile.head(30, multiple_iterators=False)]
print('Flags for some reads (first 30 reads): ' + str(some_read_flags))
unmapped_reads = 0
number_of_reads = 0
number_of_reads_zero_quality = 0
avg_map_quality = 0
avg_non_zero_map_quality = 0
for read in samfile:
    number_of_reads += 1
    avg_map_quality += read.mapping_quality
    if(read.is_unmapped):
        unmapped_reads +=1;
    if(read.mapping_quality == 0):
        number_of_reads_zero_quality +=1
    else:
        avg_non_zero_map_quality += read.mapping_quality
avg_map_quality /= number_of_reads
avg_non_zero_map_quality = avg_non_zero_map_quality/(number_of_reads - number_of_reads_zero_quality)
print('Total number of reads: ' + str(number_of_reads))
print('Total number of unmapped reads: ' + str(unmapped_reads))
print('Total number of reads with zero mapping quality: ' + str(number_of_reads_zero_quality))
print('Average mapping quality for all reads: ' + str(avg_map_quality))
print('Average mapping quality for reads with mapping quality greater than zero: ' + str(avg_non_zero_map_quality))
```
